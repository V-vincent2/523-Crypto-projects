## Precis
"Momentum effects in the cryptocurrency market after one-day abnormal returns" by Guglielmo Maria Caporale and Alex Plastun was published in the Financial Markets and Portfolio Management journal in 2020. The authors examine the momentum effects same-day and following one-day abnormal returns in the cryptocurrency market. There are 3 major points in the paper: 1. There were significant differences in hourly returns on days with abnormal returns compared to average days, indicating momentum effects on those days. 2. Abnormal returns can be detected before the day's end, allowing room for trading strategies. 3. A continuation of the momentum effect into the following days, providing further opportunities for profit. Tying all together, the authors conclude that these patterns of momentum following abnormal returns present exploitable opportunities for profit. The findings are valuable in the academic world for momentum research of the cryptocurrencies and profitable trading strategies in the cryptocurrency market.

The first point presented by Caporale and Plastun is the exploration of the momentum effects  in the cryptocurrency market on the days marked by abnormal returns. To test this hypothesis, they first defined abnormal returns as returns that are 2 SD higher than normal hourly returns in a time period for Bitcoin, and 1.5 SD for Ethereum and Litecoin. The data they used are from January 1, 2015, to September 1, 2019 for the hourly returns on these three coins against the US dollar. To see the effects of the abnormal, they used cumulative abnormal returns to see changes after they recorded abnormal returns on that day. The findings demonstrate that on days following the identification of abnormal returns(positive/negative), the cryptocurrency prices show patterns of continuing moving in the same direction as the abnormal return(positive/negative), thereby showing the momentum effects. This observation crucially suggests that the cryptocurrency market displays patterns that can be anticipated for the same trading day, challenging the traditional Efficient Market Hypothesis.

After demonstrating the momentum effects after abnormal hourly returns, the paper furthermore shows the pattern of the abnormal returns and the momentum effects, allowing investors to potentially capitalize on certain strategies. The paper identified the usual time for the abnormal returns of three coins which are 18:00, 13:00, 12:00 for the positive abnormal returns and 16:00, 14:00, 13:00 for the negative abnormal returns. These times show that there are plenty of rooms to execute strategies for same day profits. The authors then proceed to test out the strategy 1: After realizing the current day is abnormal(after the time), open a position in the direction of abnormal returns and close at the end of the day. The results show a 90% successful trade average for positive abnormal return days and 85% successful trade average for negative days and returns that are much higher than usual returns of US investments(>143% profit). The conclusion drawn here is that the cryptocurrency market offers detectable and exploitable opportunities for profit within trading hours of the same day and strategies are proven to work.

As Caporale and Plastun identify strategy for same day momentum trading, they also include exploration and testing for the day after abnormal return day.  For the day after, it is a bit complicated since the momentum effect did not stay for some cases but instead the contrarian effect takes in for some cases, notably positive Bitcoin and negative Ethereum. However, there are still momentum effects for the others, and contrarian effects can be exploited too. Therefore they came up with strategy 2: At the beginning of the day after the abnormal returns, open a position in the direction of the abnormal returns and close on the basis of the timing parameters for the momentum effect. If this effect is not present, a contrarian trading strategy should be used which is the opposite direction of this strategy. For the results, Strategy 2 is less profitable: the average number of successful trades is less than 50% and only in 1 out of 3 cases there are statistically significant differences from random trading, but all the simulations generate profits which on an annual basis are much higher than average returns on investment in the USA as well. This shows the momentum effect/contrarian effect can be extended to the second day but just not as effective. This third point reinforced the second point and provided more room for exploitable opportunities the day after abnormal returns.

This paper offers evidence of momentum effects in the cryptocurrency market day of and the following day of abnormal returns, and it is a great stepping stone between the traditional momentum strategy and modern world asset class. By detailing specific strategies that capitalize on these effects, Caporale and Plastun contribute to academic discourse on market efficiency, and they also provide practical tools for traders. Future research could extend this analysis to other markets or assets and explore the underlying causes of momentum effects.

```python
#write the function for fetching data for three coins
def fetch_crypto_data(sym, timeStart, timeEnd, barSize="3600"):
    apiUrl = "https://api.pro.coinbase.com"
    delta = timedelta(seconds=int(barSize))
    all_data = []

    # Loop to fetch data in chunks
    while timeEnd > datetime(2015, 1, 1, 0, 0, 0, 0):
        parameters = {
            "start": timeStart.isoformat(),
            "end": timeEnd.isoformat(),
            "granularity": barSize,
        }
        response = requests.get(f"{apiUrl}/products/{sym}/candles", params=parameters)
        data = response.json()
        
        # Break if no data is returned
        if not data or response.status_code != 200:
            break
        
        all_data.extend(data)
        # Update the end time to the earliest time in the fetched data
        timeEnd = timeStart
        # Update the start time for the next loop iteration
        timeStart -= (300 * delta)
        
    # Convert the collected data into a DataFrame
    df = pd.DataFrame(all_data, columns=["time", "low", "high", "open", "close", "volume"])
    df["date"] = pd.to_datetime(df["time"], unit='s')
    df.sort_values("date", inplace=True)
    df.reset_index(drop=True, inplace=True)
    return df[["date", "open", "high", "low", "close", "volume"]]

# Initial parameters
barSize = "3600"  # 1 hour
timeEnd = datetime(2024, 4, 18, 0, 0, 0, 0)
timeStart = timeEnd - (300 * timedelta(seconds=int(barSize)))

# Fetch data for each cryptocurrency
btc_df = fetch_crypto_data("BTC-USD", timeStart, timeEnd, barSize)
ltc_df = fetch_crypto_data("LTC-USD", timeStart, timeEnd, barSize)
eth_df = fetch_crypto_data("ETH-USD", timeStart, timeEnd, barSize)

#thanks to help from discussion board
```

### Abnormal Returns Defination
In the paper, there are specific ways to define abnormal returns. $R_i$ is defined within the context of abnormal thresholds:

$$ R_i < \overline{R}_n - k \times \delta_n, $$

where $k$ is the number of standard deviations used to identify the abnormal returns ($k = 2$ for BTCUSD and $k = 1.5$ for ETHUSD and LTCUSD), $k$ being chosen based on the sample size to generate in each case a sufficient number of detected abnormal returns); $\overline{R}_n$ is the average size of daily returns for period $n$:

$$ \overline{R}_n = \frac{\sum_{i=1}^{n} R_i}{n}, $$

and $\delta_n$ is the standard deviation of daily returns for period $n$:

$$ \delta_n = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (R_i - \overline{R}_n)^2}. $$

However, given this equation is not enough, since no where in the paper it says what is n for the period, and in reality we don't know if the authors actually used data that is before the observation price change. The paper did have reference to another paper which says from 30 - 60 is best. I have used a function located at the end of this document that helped me to find the best number to match the original paper.


```python
#This point we find the days that are Abnormal returns

# Parameters
k = 2  # Number of sd for AR for BTC
n_btc = 86
#number of observations
#Hourly


# Rolling mean and sd
btc_final_hourly_df['rolling_mean_hr']= btc_final_hourly_df['hourly_returns'].rolling(window=n_btc).mean()
btc_final_hourly_df['rolling_std_hr'] = btc_final_hourly_df['hourly_returns'].rolling(window=n_btc).std()


# AR identify
btc_final_hourly_df['abnormal_return_hr'] = btc_final_hourly_df.apply(
    lambda row: 'abnormal' if abs(row['hourly_returns'] - row['rolling_mean_hr']) > k * row['rolling_std_hr'] else 'normal', axis=1)

#Daily
# Rolling mean and sd
btc_final_daily_df['rolling_mean_dy'] = btc_final_daily_df['daily_returns'].rolling(window=n_btc).mean()
btc_final_daily_df['rolling_std_dy'] = btc_final_daily_df['daily_returns'].rolling(window=n_btc).std()
btc_final_daily_df['mean_dy'] = btc_final_daily_df['daily_returns'].mean()
btc_final_daily_df['std_dy'] = btc_final_daily_df['daily_returns'].std()
btc_final_daily_df['expanding_mean_dy'] = btc_final_daily_df['daily_returns'].expanding(min_periods=1).mean()
btc_final_daily_df['expanding_std_dy'] = btc_final_daily_df['daily_returns'].expanding(min_periods=1).std()

# AR identify
btc_final_daily_df['abnormal_return_dy'] = btc_final_daily_df.apply(
    lambda row: 'abnormal' if abs(row['daily_returns'] - row['rolling_mean_dy']) > k * row['rolling_std_dy'] else 'usual', axis=1)
```

```python
def compute_crypto_statistics(daily_df, hourly_df):
    # Make copies to avoid modifying the original dataframes
    daily_df = daily_df.copy()
    hourly_df = hourly_df.copy()

    # descriptive statistics function
    def descriptive_stats(data, value_column):
        stats = {
            'Mean': data[value_column].mean(),
            'Standard Error': data[value_column].sem(),
            'Median': data[value_column].median(),
            'Mode': 0,
            'Std. Dev.': data[value_column].std(),
            'Variance': data[value_column].var(),
            'Kurtosis': data[value_column].kurtosis(),
            'Skewness': data[value_column].skew(),
            'Minimum': data[value_column].min(),
            'Maximum': data[value_column].max(),
            'Sum': data[value_column].sum(),
            'Observations': data[value_column].count()
        }
        return stats

    # Extract abnormal dates
    abnormal_dates = daily_df[daily_df['abnormal_return_dy'] == 'abnormal']['date_only']
    hourly_df['day_type'] = hourly_df['date_only'].isin(abnormal_dates)
    hourly_df['day_type'] = hourly_df['day_type'].replace({True: 'abnormal', False: 'usual'})

    #find all the postive days from daily returns
    all_positive_days = daily_df[daily_df['daily_returns'] > 0]['date'].dt.date

    #then find all the abnormal postive days and find their hourly returns
    positive_days_hourly_data = hourly_df[hourly_df['date'].dt.date.isin(all_positive_days)]
    abnormal_positive_hourly = positive_days_hourly_data[positive_days_hourly_data['day_type'] == 'abnormal']

    # same for negative
    all_negative_days = daily_df[daily_df['daily_returns'] < 0]['date'].dt.date

    # hourly returns for the abnormal
    negative_days_hourly_data = hourly_df[hourly_df['date'].dt.date.isin(all_negative_days)]
    abnormal_negative_hourly = negative_days_hourly_data[negative_days_hourly_data['day_type'] == 'abnormal']

    # Calculate the statistics
    stats = {
        'Daily Data': descriptive_stats(daily_df, 'daily_returns'),
        'Positive Abnormal Days': descriptive_stats(daily_df[(daily_df['daily_returns'] > 0) & 
                        (daily_df['abnormal_return_dy'] == 'abnormal')], 'daily_returns'),
        'Negative Abnormal Days': descriptive_stats(daily_df[(daily_df['daily_returns'] < 0) & 
                        (daily_df['abnormal_return_dy'] == 'abnormal')], 'daily_returns'),
        'Hourly Data': descriptive_stats(hourly_df, 'hourly_returns'),
        'Positive Abnormal Hours': descriptive_stats(abnormal_positive_hourly, 'hourly_returns'),
        'Negative Abnormal Hours': descriptive_stats(abnormal_negative_hourly, 'hourly_returns'),  
    }

    #convert to df 
    stats_df = pd.DataFrame(stats)

    return stats_df
btc_stats_df = compute_crypto_statistics(btc_daily_data_up_to_2019, btc_hourly_data_up_to_2019)
btc_stats_df
```

```python
def compute_hourly_ttest_pos_day(daily_df, hourly_df):
    # Make copies to avoid modifying the original dataframes
    daily_df = daily_df.copy()
    hourly_df = hourly_df.copy()
    
    # Extract abnormal dates
    abnormal_dates = daily_df[daily_df['abnormal_return_dy'] == 'abnormal']['date_only']
    hourly_df['day_type'] = hourly_df['date_only'].isin(abnormal_dates)
    hourly_df['day_type'] = hourly_df['day_type'].replace({True: 'abnormal', False: 'usual'})
    all_positive_days = daily_df[daily_df['daily_returns'] > 0]['date'].dt.date

    positive_days_hourly_data = hourly_df[hourly_df['date'].dt.date.isin(all_positive_days)]
    abnormal_positive_hourly = positive_days_hourly_data[positive_days_hourly_data['day_type'] == 'abnormal']
    usual_positive_hourly = positive_days_hourly_data[positive_days_hourly_data['day_type'] != 'abnormal']
    
    # Initialize list to store the data
    hourly_stats_data = []
    
    for hour in range(24):
        # Filter the hourly data for the specific hour
        hourly_data_abnormal = abnormal_positive_hourly[abnormal_positive_hourly['date'].dt.hour == hour]
        hourly_data_usual = usual_positive_hourly[usual_positive_hourly['date'].dt.hour == hour]
        
        # Calculate average returns and standard deviations
        avg_return_abnormal = hourly_data_abnormal['hourly_returns'].mean() * 100
        std_dev_abnormal = hourly_data_abnormal['hourly_returns'].std() * 100
        avg_return_usual = hourly_data_usual['hourly_returns'].mean() * 100
        std_dev_usual = hourly_data_usual['hourly_returns'].std() * 100
        
        # t-test
        t_stat = None
        if len(hourly_data_abnormal) > 1 and len(hourly_data_usual) > 1:
            t_stat, _ = ttest_ind(hourly_data_abnormal['hourly_returns'], hourly_data_usual['hourly_returns'], equal_var=False)
            t_stat = round(t_stat, 2)
        
        # Append the data to the list
        hourly_stats_data.append({
            'Hour': f"{hour}:00",
            'Average Return (OD)': f"{avg_return_abnormal:.2f}%" if avg_return_abnormal is not pd.NA else "N/A",
            'Std Dev (OD)': f"{std_dev_abnormal:.2f}%" if std_dev_abnormal is not pd.NA else "N/A",
            'Observations (OD)': len(hourly_data_abnormal),
            'Average Return (UD)': f"{avg_return_usual:.2f}%" if avg_return_usual is not pd.NA else "N/A",
            'Std Dev (UD)': f"{std_dev_usual:.2f}%" if std_dev_usual is not pd.NA else "N/A",
            'Observations (UD)': len(hourly_data_usual),
            't criterion': t_stat if t_stat is not None else "N/A"
        })
    
    # Convert the list to a DataFrame
    return pd.DataFrame(hourly_stats_data)

# Usage
btc_positive_day_table = compute_hourly_ttest_pos_day(btc_daily_data_up_to_2019, btc_hourly_data_up_to_2019)
btc_positive_day_table
```

### Cumulative Abnonal return  and Hypothesis Two
The cumulative abnormal returns approach is based on MacKinlay and Richardson (1991) and is standard for event studies. Abnormal returns are defined as follows:

$$ AR_t = R_t - E(R_t), $$

where $R_t$ is the return at time $t$ and $E(R_t)$ is the corresponding average return computed over the whole sample period as follows:

$$ E(R_t) = \frac{1}{T} \sum_{i=1}^{T} R_i,  $$

where $T$ is the sample size, and

The cumulative abnormal return, denoted as $CAR_i$, is simply the sum of the abnormal returns:

$$ CAR_i = \sum_{i=1}^{24} AR_i,  $$

where $i$ starts with 1 (the first hour of the trading day) and ends with 24 (the last hour of the trading day). A day consists of 24 hours.

However, here the authors again didn't specify the N and sample range, but from their supplementary materials I was able to deduct that they have used the whole sample to find the standard deviation and mean to find the cut-off of the timing of abnormal returns or the crossover which means if they are using this for prediction, there is a flaw to the strategy. 

```python

def calculate_ar_car_cross(positive_day_df, negative_day_df, stats_df, k):
    # Calculate the starting cross value based on daily mean and standard deviation
    #table for AR, CAR, and cross AR timing BTC day of
    mean_daily = stats_df['daily_returns'].mean()
    std_dev_daily = stats_df['daily_returns'].std()
    
    starting_cross_value_pos = mean_daily + k * std_dev_daily
    starting_cross_value_neg = mean_daily - k * std_dev_daily
    positive_day_df = positive_day_df.copy()
    negative_day_df = negative_day_df.copy()
    
    # Convert percentage strings to floats for calculations from table
    positive_day_df['Average Return (OD)'] = positive_day_df['Average Return (OD)'].str.rstrip('%').astype('float') / 100.0
    negative_day_df['Average Return (OD)'] = negative_day_df['Average Return (OD)'].str.rstrip('%').astype('float') / 100.0
    positive_day_df['Average Return (UD)'] = positive_day_df['Average Return (UD)'].str.rstrip('%').astype('float') / 100.0
    negative_day_df['Average Return (UD)'] = negative_day_df['Average Return (UD)'].str.rstrip('%').astype('float') / 100.0
    
    # Initialize the DataFrame with AR values
    ar_car_cross_df = pd.DataFrame({
        'Hour': positive_day_df['Hour'],
        'AR Positive': positive_day_df['Average Return (OD)'] - positive_day_df['Average Return (UD)'],
        'CAR Positive': np.nan,  # Placeholder for CAR Positive values
        'Cross Positive': starting_cross_value_pos,  # Placeholder for Cross Positive values
        'AR Negative': negative_day_df['Average Return (OD)'] - negative_day_df['Average Return (UD)'],
        'CAR Negative': np.nan,  # Placeholder for CAR Negative values
        'Cross Negative': starting_cross_value_neg  # Placeholder for Cross Negative values
    })

    # Calculate CAR and Cross values iteratively for each hour
    for i in range(len(ar_car_cross_df)):
        ar_car_cross_df.loc[i, 'CAR Positive'] = ar_car_cross_df['AR Positive'].iloc[:i+1].sum()
        ar_car_cross_df.loc[i, 'CAR Negative'] = ar_car_cross_df['AR Negative'].iloc[:i+1].sum()
        if i == 0:
            ar_car_cross_df.loc[i, 'Cross Positive'] = starting_cross_value_pos - positive_day_df.loc[i, 'Average Return (OD)']
            ar_car_cross_df.loc[i, 'Cross Negative'] = starting_cross_value_neg - negative_day_df.loc[i, 'Average Return (OD)']
        else:
            ar_car_cross_df.loc[i, 'Cross Positive'] = ar_car_cross_df.loc[i-1, 'Cross Positive'] - positive_day_df.loc[i, 'Average Return (OD)']
            ar_car_cross_df.loc[i, 'Cross Negative'] = ar_car_cross_df.loc[i-1, 'Cross Negative'] - negative_day_df.loc[i, 'Average Return (OD)']
    
    return ar_car_cross_df

# Usage
BTC_day_ar_car_cross_df = calculate_ar_car_cross(btc_positive_day_table, btc_negative_day_table, btc_daily_data_up_to_2019,2)
BTC_day_ar_car_cross_df
```

```python
def compute_day_after_ttest_pos(daily_df, hourly_df):

    daily_df = daily_df.copy()
    hourly_df = hourly_df.copy()

    # Extract dates for positive abnormal returns and their following days
    abnormal_dates = daily_df[daily_df['abnormal_return_dy'] == 'abnormal']['date_only']
    daily_df['day_type'] = daily_df['date_only'].isin(abnormal_dates)
    daily_df['day_type'] = daily_df['day_type'].replace({True: 'abnormal', False: 'usual'})


    all_positive_days_abnormal = daily_df[(daily_df['daily_returns'] > 0 )&(daily_df['day_type'] == 'abnormal') ]['date'].dt.date
    day_after_positive_abnormal = all_positive_days_abnormal + pd.Timedelta(days=1)
    positive_after_hourly_data = hourly_df[hourly_df['date'].dt.date.isin(day_after_positive_abnormal)]
    

    # Initialize list for results
    day_after_data = []

    # Calculate statistics and t-tests for each hour
    for hour in range(24):
        hourly_data = positive_after_hourly_data[positive_after_hourly_data['date'].dt.hour == hour]
        avg_return = hourly_data['hourly_returns'].mean() * 100  # Convert to percentage
        std_dev = hourly_data['hourly_returns'].std() * 100  # Convert to percentage
        num_observations = hourly_data.shape[0]

        # Calculate stats for usual hourly data
        usual_hourly_data = hourly_df[hourly_df['date'].dt.hour == hour]
        avg_return_usual = usual_hourly_data['hourly_returns'].mean() * 100  # Convert to percentage
        std_dev_usual = usual_hourly_data['hourly_returns'].std() * 100  # Convert to percentage

        # Perform t-test
        t_stat = None
        if num_observations > 1 and not usual_hourly_data.empty:
            t_stat, _ = ttest_ind(hourly_data['hourly_returns'], usual_hourly_data['hourly_returns'], equal_var=False)
            t_stat = round(t_stat, 2)  # Round the t-statistic

        # Append results
        day_after_data.append({
            'Hour': f"{hour}:00",
            'Average Return (OD)': f"{avg_return:.2f}%",
            'Std Dev (OD)': f"{std_dev:.2f}%",
            'Observations (OD)': num_observations,
            'Average Return (UD)': f"{avg_return_usual:.2f}%",
            'Std Dev (UD)': f"{std_dev_usual:.2f}%",
            'Observations (UD)': len(usual_hourly_data),
            't criterion': t_stat if t_stat is not None else 'N/A'
        })

    # Convert the list to a DataFrame
    return pd.DataFrame(day_after_data)


# Usage
pos_btc_day_after_df = compute_day_after_ttest_pos(btc_daily_data_up_to_2019, btc_hourly_data_up_to_2019)
pos_btc_day_after_df
```

```python
def calculate_ar_car_cross_after(positive_day_df, negative_day_df):

    positive_day_df = positive_day_df.copy()
    negative_day_df = negative_day_df.copy()
    
    # Convert percentage strings to floats for calculations from table
    positive_day_df['Average Return (OD)'] = positive_day_df['Average Return (OD)'].str.rstrip('%').astype('float') / 100.0
    negative_day_df['Average Return (OD)'] = negative_day_df['Average Return (OD)'].str.rstrip('%').astype('float') / 100.0
    positive_day_df['Average Return (UD)'] = positive_day_df['Average Return (UD)'].str.rstrip('%').astype('float') / 100.0
    negative_day_df['Average Return (UD)'] = negative_day_df['Average Return (UD)'].str.rstrip('%').astype('float') / 100.0
    
    # Initialize the DataFrame with AR values
    ar_car_cross_df = pd.DataFrame({
        'Hour': positive_day_df['Hour'],
        'AR Positive': positive_day_df['Average Return (OD)'] - positive_day_df['Average Return (UD)'],
        'CAR Positive': np.nan,  # Placeholder for CAR Positive values
        'AR Negative': negative_day_df['Average Return (OD)'] - negative_day_df['Average Return (UD)'],
        'CAR Negative': np.nan  # Placeholder for CAR Negative values
    })

    # Calculate CAR and Cross values iteratively for each hour
    for i in range(len(ar_car_cross_df)):
        ar_car_cross_df.loc[i, 'CAR Positive'] = ar_car_cross_df['AR Positive'].iloc[:i+1].sum()
        ar_car_cross_df.loc[i, 'CAR Negative'] = ar_car_cross_df['AR Negative'].iloc[:i+1].sum()
    return ar_car_cross_df

# Usage
BTC_after_ar_car_cross_df = calculate_ar_car_cross_after(pos_btc_day_after_df, neg_btc_day_after_df)
BTC_after_ar_car_cross_df
```

## Strategy 1 testing, Support for Hypothesis Two

Here we have strategy one, When it becomes clear that the current day is an abnormal returns day a position in the direction of abnormal returns should be opened. This position should then be closed at the end of
the day.

However, the authors again didn't include all the parameters and details of the process. It is unclear to tell how the current day is "abnormal" whether it is past a threshold or use the timing of abnormal return from earlier. If the the authors use the timing from earlier that would be flawed since that was collected from the data as a whole which shouldn't be used for strategy or prediction. In this paper for replication, I thought about the rolling mean and rolling standard deviations as threshold for each day to see if the return has met "abnormal" then follow the trade.

```python
#strategy one testing
def strategy_1_evaluation_pos(daily_df, hourly_df, k=2, name = 'BTCUSD'):
    """
    Evaluates a trading strategy that opens positions dynamically based on rolling mean and sd

    Parameters:
    - daily_df (DataFrame): DataFrame containing daily data with abnormal return labels.
    - hourly_df (DataFrame): DataFrame containing hourly data.
    - k (int): Number of standard deviations above the mean to set as threshold.

    Returns:
    - DataFrame: Summary of the trading strategy's performance.
    """
    # Prepare data
    daily_df = daily_df.copy()
    daily_df['date_only'] = pd.to_datetime(daily_df['date']).dt.date
    hourly_df = hourly_df.copy()
    hourly_df['date'] = pd.to_datetime(hourly_df['date'])

    # Define threshold for each day
    daily_df['threshold'] = daily_df['rolling_mean_dy'] + k * daily_df['rolling_std_dy']

    results = []
    profits = []

    for _, row in daily_df.iterrows():
        if row['abnormal_return_dy'] == 'abnormal' and row['daily_returns'] > 0:  # Positive abnormal return day
            trade_date = row['date_only']
            threshold = row['threshold']
            day_hourly_data = hourly_df[hourly_df['date'].dt.date == trade_date].copy()

            if day_hourly_data.empty:
                continue

            # Calculate cumulative returns from 00:00 and find the opening hour
            day_hourly_data['cumulative_return'] = day_hourly_data['hourly_returns'].cumsum()
            opening_hour_data = day_hourly_data[day_hourly_data['cumulative_return'] > threshold]

            if not opening_hour_data.empty:
                opening_hour = opening_hour_data.iloc[0]['date']
                opening_price = opening_hour_data.iloc[0]['close']
                closing_price = day_hourly_data.iloc[-1]['close']

                # Calculate profit
                profit = (closing_price - opening_price) / opening_price * 100
                profits.append(profit)

                results.append({
                    'Date': trade_date,
                    'Open Hour': opening_hour,
                    'Open Price': opening_price,
                    'Close Price': closing_price,
                    'Profit': profit
                })

    results_df = pd.DataFrame(results)
    total_trades = len(results_df)
    successful_trades = len(results_df[results_df['Profit'] > 0])
    total_profit = sum(profits)
    profit_per_year = total_profit / (daily_df['date'].dt.year.nunique())
    profit_per_trade = np.mean(profits) if profits else 0

    # Perform t-test against zero (no skill trading)
    t_stat, p_value = ttest_1samp(profits, 0) if profits else (None, None)

    # Prepare summary
    summary_df = pd.DataFrame({
        'Instrument': [name],
        'Number of trades': [total_trades],
        'Number of successful trades': [successful_trades],
        'Successful trades %': [100 * successful_trades / total_trades if total_trades > 0 else 0],
        'Profit %': [total_profit],
        'Profit % per year': [profit_per_year],
        'Profit % per trade': [profit_per_trade],
        't-test value': [t_stat],
        't-test status': ['Rejected' if p_value < 0.05 else 'Not Rejected']
    })

    return summary_df

btc_strategy_stats = strategy_1_evaluation_pos(btc_daily_data_up_to_2019, btc_hourly_data_up_to_2019,2,'BTCUSDP')
btc_strategy_stats
```

## Strategy 2 testing, Support for Hypothesis Three


Strategy 2 At the beginning of the day after the abnormal returns, a position in the direction of the abnormal returns should be opened. This position should then be closed on the basis of the timing parameters for the momentum effect. If this effect is not present, a contrarian trading strategy should be used: at thebeginning of the day after the abnormal returns a position in the opposite direction to the abnormal returns should be opened.

```python
def strategy_two_evaluation(daily_df, hourly_df, trade_type='momentum', day_type='positive',close_hour='21:00',name= 'BTCUSD'):
    """
    Parameters:
    - daily_df (DataFrame): DataFrame containing daily data with abnormal return labels.
    - hourly_df (DataFrame): DataFrame containing hourly data.
    - trade_type (str): 'momentum' or 'contrarian' to specify the trade direction strategy.
    - day_type (str): 'positive' or 'negative' to specify the day type after which trades are made.
    - close_hour (datetime.time): Specific hour to close trades; determined dynamically if None.
    
    Returns:
    - DataFrame: Summary of the trading strategy's performance.
    """
    daily_df = daily_df.copy()
    hourly_df = hourly_df.copy()
    daily_df['date'] = pd.to_datetime(daily_df['date']).dt.date
    hourly_df['date'] = pd.to_datetime(hourly_df['date'])

    
    # Selecting the abnormal days based on day_type
    if day_type == 'positive':
        target_days = daily_df[(daily_df['abnormal_return_dy'] == 'abnormal') & (daily_df['daily_returns'] > 0)].copy()
    else:
        target_days = daily_df[(daily_df['abnormal_return_dy'] == 'abnormal') & (daily_df['daily_returns'] < 0)].copy()

    target_days['next_day'] = target_days['date'] + pd.Timedelta(days=1)

    # Filter hourly data for next days
    relevant_hourly_data = hourly_df[hourly_df['date'].dt.date.isin(target_days['next_day'])]
    
    profits = []
    for _, day_row in target_days.iterrows():
        day_data = relevant_hourly_data[relevant_hourly_data['date'].dt.date == day_row['next_day']]
        if not day_data.empty:
            open_price = day_data.iloc[0]['close']
                
            close_data = day_data[day_data['date'].dt.time == pd.to_datetime(close_hour).time()]
            if not close_data.empty:
                close_price = close_data['close'].values[0]
            if (trade_type == 'momentum' and day_type == 'positive') or (trade_type == 'contrarian' and day_type == 'negative'):
                profit = (close_price - open_price) / open_price * 100
            else:  
                profit = (open_price - close_price) / open_price * 100
            
            profits.append(profit)

    # Compute overall strategy statistics
    total_trades = len(profits)
    successful_trades = sum(1 for x in profits if x > 0)
    total_profit = sum(profits)
    profit_per_trade = total_profit / total_trades if total_trades > 0 else 0
    annual_profit = total_profit / ((daily_df['date'].max() - daily_df['date'].min()).days / 365.25)
    success_rate = successful_trades / total_trades * 100 if total_trades > 0 else 0

    # Statistical Test
    t_stat, p_value = ttest_1samp(profits, 0) if profits else (None, None)
    t_test_status = 'Not rejected' if p_value >= 0.05 else 'Rejected'

    # Summary DataFrame
    summary_df = pd.DataFrame({
        'Instrument': [name],
        'Number of trades': [total_trades],
        'Number of successful trades': [successful_trades],
        'Successful trades %': [success_rate],
        'Profit %': [total_profit],
        'Profit % per year': [annual_profit],
        'Profit % per trade': [profit_per_trade],
        't-test value': [t_stat],
        't-test status': [t_test_status]
    })

    return summary_df


strategy_stats_2_btc_pos = strategy_two_evaluation(btc_daily_data_up_to_2019, btc_hourly_data_up_to_2019,
                                         trade_type='momentum', day_type='positive', close_hour='14:00',name= 'BTCUSDP')
strategy_stats_2_btc_pos
```

```python
#this is the function that I used to find n_days
def find_best_n_days(daily_df, hourly_df, start=3, end=110):
    """
    Iterates over a range of n_days values to find the one that yields the maximum number of successful trades.

    Parameters:
    - daily_df (DataFrame): DataFrame containing daily data with abnormal return labels.
    - hourly_df (DataFrame): DataFrame containing hourly data.
    - start (int): Starting value of n_days to test.
    - end (int): Ending value of n_days to test.

    Returns:
    - int: The best n_days value.
    - DataFrame: Statistics of the strategy performance using the best n_days.
    """
    best_n_days = start
    max_successful_trades = 0
    best_stats = pd.DataFrame()

    for i in range(start, end + 1):
        strategy_stats = strategy_1_evaluation_neg(daily_df, hourly_df, n_days=i)
        
        # Check the number of successful trades
        current_successful_trades = strategy_stats['t-test value'].item()
        
        if current_successful_trades > max_successful_trades:
            max_successful_trades = current_successful_trades
            best_n_days = i
            best_stats = strategy_stats

    return best_n_days, best_stats

# Example usage:
#best_n_days, best_stats = find_best_n_days(btc_daily_data_up_to_2019, btc_hourly_data_up_to_2019)
#print("Best n_days:", best_n_days)
#print("Strategy statistics using the best n_days:")
#print(best_stats)
#86 was best number of days
```
