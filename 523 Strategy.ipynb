In this new strategy project, I aim to fix the above flaws and see the results of this extensive research and development.

Since crypto currencies are traded 24 hours a day and 7 days a week around the world, it is hard to argue what time is the “beginning” or “ending” of the day since every day is overlapped in Asia, Europe, and America time zones. So using UTC timezone and assuming the momentum continues until the end of the UTC hours could be revised to an alternative hypothesis which is “After an hour of abnormal returns (both positive and negative), there is a momentum (or contrarian) effect in cryptocurrency prices in the short term.” Here, I replaced the daily abnormal return to hour(s) of abnormal return to replace the concept of days.This formulates the first hypothesis. The idea and the general structure would be the same as the methods in the original replication paper, but the momentum and contrarian periods would be reevaluated and so as definitions of abnormal returns.

In the replication paper, the signal of abnormal return of that day is whenever the cumulative hourly return of that day is greater than the mean of daily return + k Standard deviation of the daily return across all testing periods. This method for signaling abnormal return is not efficient as it summarizes the entire data to two variables and it is also flawed since it is using future data for current trade/strategy. Instead in this strategy paper, I am using moving average and k standard deviations from the past n periods for the signal, and there will be different phases for different standard deviations to use which will be explained in the second hypothesis. 

For the second hypothesis, its goal is to fix the flaw of volatility clustering observed in cryptocurrencies. (Li, Y., Urquhart, A., Wang, P., & Zhang, W. (2021)). Therefore hypothesis two is needed to enhance the momentum and contrarian model:  The momentum/contrarian effect varies over time due to the volatility clustering behavior of these assets resulting in high volatility phases vs. stable phases. The parameters and threshold for deciding high volatility phases vs stable phases will be decided in the training set of the data using Average True Range (ATR) and GARCH-type models. The high-volatility phase will have a different standard-deviation than the stable phases when determining the abnormal returns. 

The two new hypotheses and the revised methods will be the aim of this strategy project as its goal is to find the momentum/contrarian effect more accurately and efficiently. It is not expected to show better results of profitability over the original paper but to show the possibilities of different perspectives as well as new methods to the same problem.

Abnormal Returns: Calculate hourly returns that significantly deviate from a moving average. (Different than replication project)

Trigger Signal: An abnormal return that is beyond a predefined threshold ( k standard deviations from a moving average). The trigger signal will be different for high volatility phases vs. stable phases.

```python
#write the function for fetching data for three coins
def fetch_crypto_data(sym, timeStart, timeEnd, barSize="3600"):
    apiUrl = "https://api.pro.coinbase.com"
    delta = timedelta(seconds=int(barSize))
    all_data = []

    # Loop to fetch data in chunks
    while timeEnd > datetime(2019, 1, 1, 0, 0, 0, 0):
        parameters = {
            "start": timeStart.isoformat(),
            "end": timeEnd.isoformat(),
            "granularity": barSize,
        }
        response = requests.get(f"{apiUrl}/products/{sym}/candles", params=parameters)
        data = response.json()
        
        # Break if no data is returned
        if not data or response.status_code != 200:
            break
        
        all_data.extend(data)
        # Update the end time to the earliest time in the fetched data
        timeEnd = timeStart
        # Update the start time for the next loop iteration
        timeStart -= (300 * delta)
        
    # Convert the collected data into a DataFrame
    df = pd.DataFrame(all_data, columns=["time", "low", "high", "open", "close", "volume"])
    df["date"] = pd.to_datetime(df["time"], unit='s')
    df.sort_values("date", inplace=True)
    df.reset_index(drop=True, inplace=True)
    return df[["date", "open", "high", "low", "close", "volume"]]

# Initial parameters
barSize = "3600"  # 1 hour
timeEnd = datetime(2024, 5, 20, 0, 0, 0, 0)
timeStart = timeEnd - (300 * timedelta(seconds=int(barSize)))

# Fetch data for each cryptocurrency
btc_df = fetch_crypto_data("BTC-USD", timeStart, timeEnd, barSize)
sol_df = fetch_crypto_data("SOL-USD", timeStart, timeEnd, barSize)
eth_df = fetch_crypto_data("ETH-USD", timeStart, timeEnd, barSize)
```

### Abnormal Returns 

The idea of abnormal returns is very similar to the replication project's abnormal returns. However, that one is based on daily returns and this strategy is purely on the cumalitive hourly return that is independent from days or which hour it is. 

### Replication abnormal return definition
$R_i$ is defined within the context of abnormal thresholds:

$$ R_i < \overline{R}_n - k \times \delta_n, $$

where $k$ is the number of standard deviations used to identify the abnormal returns ($k = 2$ for BTCUSD and $k = 1.5$ for ETHUSD and LTCUSD), $k$ being chosen based on the sample size to generate in each case a sufficient number of detected abnormal returns); $\overline{R}_n$ is the average size of daily returns for period $n$:

$$ \overline{R}_n = \frac{\sum_{i=1}^{n} R_i}{n}, $$

and $\delta_n$ is the standard deviation of daily returns for period $n$:

$$ \delta_n = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (R_i - \overline{R}_n)^2}. $$

The strategy above have many problems including the fixation on the daily returns. The paper concluded with momentum effects for abnormal days and the day after, but the effects on second day are not evident enough. After replication, I was firm to believe the flaw of the momentum effects day after is due to the daily structure. Therefore I have came up with a different abnormal return definition that avoids the daily structure

### New abnormal return definition

Mean and Standard Deviation of the rolling window for hours:
$$
\mu_t = \frac{1}{n} \sum_{i=t-n+1}^{t} R_i
$$

$$
\sigma_t = \sqrt{\frac{1}{n-1} \sum_{i=t-n+1}^{t} (R_i - \mu_t)^2}
$$

Calculation of the cumulative return in the look back window: 
$$
C_t = \sum_{i=t-\text{lookback}+1}^{t} R_i
$$

Threshold for postive and negative returns:
$$
\text{Threshold}_t = \mu_t + k \cdot \sigma_t \cdot \sqrt{\text{lookback}}
$$
$$
\text{Threshold}_t = \mu_t - k \cdot \sigma_t \cdot \sqrt{\text{lookback}}
$$

The look back and holding hours allow the momentum to be observed anytime instead of from the beginning of the day.

The indicator align with our hypothesis 1: After hour(s) of abnormal returns (both positive and negative), there is a momentum (or contrarian) effect in cryptocurrency prices in the short term. In order to test this indicator, we should look at how many abnormal return we would have from the data set and their distributions.

## Signal 

For the signals of this part of the project, we have the hourly momentum strategy's entering and exiting signal, and the signal for volatility which will be included in the second part of the project.

Entering signal: when the cumalitive price overcame the threshold from the indicators
$$
C_t > \text{Threshold}_t
$$
$$
C_t < \text{Threshold}_t
$$


Exit signal: target profit, stop loss or after n hours. 

Stop loss: postive momentum trading and negative contrarian trading would take long postions and negative momentum trading would take short positions.
$$
\text{Stop Loss Condition for long position: } \frac{P_{t+j} - P_t}{P_t} \leq -\text{stop loss}
$$
$$
\text{Stop Loss Condition for short position: } \frac{P_t - P_{t+j}}{P_t} \geq \text{stop loss}
$$


For the exit signal, we would take stop loss from 0.025 to 0.075 and profit target 0.025 to 0.1. Since these two exits are risk-averse strategy, we will be using them for risk management instead of a feature of them strategy so we will not be testing them. For n holding hours, we will take 10 to 30 hours to mimic the replication project of holding until next day or the day of. 

To test the entering signal, we would test the hypothesis and see the returns after the abnormal returns

## Hourly momentum/Contrarian Strategy rules


The first part of my strategy is the extension of the replication project. It aim to capture price movements after detecting abnormal returns. The abnormal returns are identified based on a threshold calculated using rolling mean and standard deviation of the returns over a specified window. 

Mean and Standard Deviation of the rolling window for hours:(Indicator)
$$
\mu_t = \frac{1}{n} \sum_{i=t-n+1}^{t} R_i
$$

$$
\sigma_t = \sqrt{\frac{1}{n-1} \sum_{i=t-n+1}^{t} (R_i - \mu_t)^2}
$$

Calculation of the cumulative return in the look back window:  (Indicator)
$$
C_t = \sum_{i=t-\text{lookback}+1}^{t} R_i
$$

Threshold for postive and negative returns:(Indicator)
$$
\text{Threshold}_t = \mu_t + k \cdot \sigma_t \cdot \sqrt{\text{lookback}}
$$

$$
\text{Threshold}_t = \mu_t - k \cdot \sigma_t \cdot \sqrt{\text{lookback}}
$$

Enter trade for postive and negative returns:(signal)
$$
C_t > \text{Threshold}_t
$$

$$
C_t < \text{Threshold}_t
$$


The trade is exited after holding for n periods or reached profit goal or if the stop loss condition is met:


Stop loss: (signal) postive momentum trading and negative contrarian trading would take long postions and negative momentum trading would take short positions.
$$
\text{Stop Loss Condition for long position: } \frac{P_{t+j} - P_t}{P_t} \leq -\text{stop loss}
$$

$$
\text{Stop Loss Condition for short position: } \frac{P_t - P_{t+j}}{P_t} \geq \text{stop loss}
$$

The idea of this hourly momentum strategy abandon the concept of days and treats every hours as standalone but also capturing the cumalitive returns over a period of time. The strategies resemble Bollinger Bands in that they use the concept of standard deviations around a moving average to identify abnormal price movements.

```python
class CombinedStrategy(bt.Strategy): #use backtrader for the strategy
    params = (
        #positive strategy parameters
        ('k_pos', 1.5),
        ('window_pos', 50),
        ('n_pos', 5),
        ('stop_loss_pos', 0.05),
        ('lookback_pos', 3),
        ('profit_goal_pos', 0.025),
        #negative strategy parameters (contrarian long)
        ('k_neg', 1.5),
        ('window_neg', 50),
        ('n_neg', 5),
        ('stop_loss_neg', 0.05),
        ('lookback_neg', 3),
        ('profit_goal_neg', 0.025),
        ('transaction_cost', 0.0025),  # 0.25%

    )

    def __init__(self):
        self.dataclose = self.data.close
        self.pct_change =self.data.hourly_returns

        # postive strat indicators
        self.rolling_mean_pos = bt.indicators.SimpleMovingAverage(self.pct_change, period=self.params.window_pos)
        self.rolling_std_pos = bt.indicators.StandardDeviation(self.pct_change, period=self.params.window_pos) * np.sqrt(self.params.lookback_pos)
        self.cumulative_returns_pos = bt.indicators.SumN(self.pct_change, period=self.params.lookback_pos)
        self.positive_threshold = self.rolling_mean_pos + self.params.k_pos * self.rolling_std_pos

        # negative strat indicators, contrarian long
        self.rolling_mean_neg = bt.indicators.SimpleMovingAverage(self.pct_change, period=self.params.window_neg)
        self.rolling_std_neg = bt.indicators.StandardDeviation(self.pct_change, period=self.params.window_neg) * np.sqrt(self.params.lookback_neg)
        self.cumulative_returns_neg = bt.indicators.SumN(self.pct_change, period=self.params.lookback_neg)
        self.negative_threshold = self.rolling_mean_neg - self.params.k_neg * self.rolling_std_neg

        self.in_trade = False
        self.trade_price = None
        self.stop_loss_price = None
        self.profit_goal_price = None
        self.entry_index = None

    def next(self):
        if self.in_trade:
            #exit conditions for long both positve and negative
            if self.dataclose[0] <= self.stop_loss_price or self.dataclose[0] >= self.profit_goal_price or len(self) - self.entry_index >= self.params.n_pos:
                self.sell()
                self.in_trade = False

        if not self.in_trade:
            # positive momentum strategy
            if self.cumulative_returns_pos[0] > self.positive_threshold[0]:
                self.buy()
                self.in_trade = True
                self.trade_price = self.dataclose[0] * (1 + self.params.transaction_cost)  #transaction cost
                self.stop_loss_price = self.trade_price * (1 - self.params.stop_loss_pos)
                self.profit_goal_price = self.trade_price * (1 + self.params.profit_goal_pos)
                self.entry_index = len(self)

            # negative contrarian (long) strategy
            elif self.cumulative_returns_neg[0] < self.negative_threshold[0]:
                self.buy()
                self.in_trade = True
                self.trade_price = self.dataclose[0] * (1 + self.params.transaction_cost)  #transaction cost
                self.stop_loss_price = self.trade_price * (1 - self.params.stop_loss_neg)
                self.profit_goal_price = self.trade_price * (1 + self.params.profit_goal_neg)
                self.entry_index = len(self)

class PandasData(bt.feeds.PandasData):
    lines = ('hourly_returns',)
    params = (('hourly_returns', 6),) 

def objective_long(trial, df):
    k_pos = trial.suggest_float('k_pos', 1.5, 2.25)
    window_pos = trial.suggest_int('window_pos', 50, 600)
    n_pos = trial.suggest_int('n_pos', 8, 30)
    stop_loss_pos = trial.suggest_float('stop_loss_pos', 0.025, 0.075)
    lookback_pos = trial.suggest_int('lookback_pos', 3, 12)
    profit_goal_pos = trial.suggest_float('profit_goal_pos', 0.025, 0.1)
    
    k_neg = trial.suggest_float('k_neg', 1.5, 2.3)
    window_neg = trial.suggest_int('window_neg', 40, 600)
    n_neg = trial.suggest_int('n_neg', 10, 30)
    stop_loss_neg = trial.suggest_float('stop_loss_neg', 0.025, 0.075)
    lookback_neg = trial.suggest_int('lookback_neg', 3, 12)
    profit_goal_neg = trial.suggest_float('profit_goal_neg', 0.02, 0.1)

    cerebro = bt.Cerebro()
    cerebro.addstrategy(CombinedStrategy, 
                        k_pos=k_pos, window_pos=window_pos, n_pos=n_pos, stop_loss_pos=stop_loss_pos, lookback_pos=lookback_pos, profit_goal_pos=profit_goal_pos,
                        k_neg=k_neg, window_neg=window_neg, n_neg=n_neg, stop_loss_neg=stop_loss_neg, lookback_neg=lookback_neg, profit_goal_neg=profit_goal_neg)
    
    data = PandasData(
        dataname=df,
        datetime='date',
        open='open',
        high='high',
        low='low',
        close='close',
        volume='volume',
        openinterest=-1,
        hourly_returns='hourly_returns'
    )
    cerebro.adddata(data)
    cerebro.broker.set_cash(100000)#common starting value
    cerebro.run()
    
    portfolio_value = cerebro.broker.get_value()
    
    if np.isnan(portfolio_value):
        return -np.inf  # Return a very low value if the portfolio value is nan
    
    return portfolio_value
```

```python
class CombinedShortingStrategy(bt.Strategy): #for the shorting side
    params = (
        # Positive strategy parameters
        ('k_pos', 1.5),
        ('window_pos', 50),
        ('n_pos', 5),
        ('stop_loss_pos', 0.05),
        ('lookback_pos', 3),
        ('profit_goal_pos', 0.025),
        
        # Negative strategy parameters (shorting)
        ('k_neg', 1.5),
        ('window_neg', 50),
        ('n_neg', 5),
        ('stop_loss_neg', 0.05),
        ('lookback_neg', 3),
        ('profit_goal_neg', 0.025),
        ('transaction_cost', 0.0025),  # 0.25%

    )

    def __init__(self):
        self.dataclose = self.data.close
        self.pct_change =self.data.hourly_returns

        # positive strat indicators
        self.rolling_mean_pos = bt.indicators.SimpleMovingAverage(self.pct_change, period=self.params.window_pos)
        self.rolling_std_pos = bt.indicators.StandardDeviation(self.pct_change, period=self.params.window_pos) * np.sqrt(self.params.lookback_pos)
        self.cumulative_returns_pos = bt.indicators.SumN(self.pct_change, period=self.params.lookback_pos)
        self.positive_threshold = self.rolling_mean_pos + self.params.k_pos * self.rolling_std_pos

        # negative strat indicators, momentum short
        self.rolling_mean_neg = bt.indicators.SimpleMovingAverage(self.pct_change, period=self.params.window_neg)
        self.rolling_std_neg = bt.indicators.StandardDeviation(self.pct_change, period=self.params.window_neg) * np.sqrt(self.params.lookback_neg)
        self.cumulative_returns_neg = bt.indicators.SumN(self.pct_change, period=self.params.lookback_neg)
        self.negative_threshold = self.rolling_mean_neg - self.params.k_neg * self.rolling_std_neg

        self.in_trade = False
        self.trade_price = None
        self.stop_loss_price = None
        self.profit_goal_price = None
        self.entry_index = None
        self.short_trade = False

    def next(self):
        if self.in_trade:
            if self.short_trade:
                # short exit for negative
                if self.dataclose[0] >= self.stop_loss_price or self.dataclose[0] <= self.profit_goal_price or len(self) - self.entry_index >= self.params.n_neg:
                    self.buy()  
                    self.in_trade = False
            else:
                # long exit for postive
                if self.dataclose[0] <= self.stop_loss_price or self.dataclose[0] >= self.profit_goal_price or len(self) - self.entry_index >= self.params.n_pos:
                    self.sell()  
                    self.in_trade = False

        if not self.in_trade:
            #positive momentum strategy
            if self.cumulative_returns_pos[0] > self.positive_threshold[0]:
                self.buy()
                self.in_trade = True
                self.trade_price = self.dataclose[0] * (1 + self.params.transaction_cost)  #transaction cost
                self.stop_loss_price = self.trade_price * (1 - self.params.stop_loss_pos)
                self.profit_goal_price = self.trade_price * (1 + self.params.profit_goal_pos)
                self.entry_index = len(self)
                self.short_trade = False

            #negative (shorting) momentum strategy
            elif self.cumulative_returns_neg[0] < self.negative_threshold[0]:
                self.sell()  #Short
                self.in_trade = True
                self.trade_price = self.dataclose[0] * (1 + self.params.transaction_cost)  #transaction cost
                self.stop_loss_price = self.trade_price * (1 + self.params.stop_loss_neg)
                self.profit_goal_price = self.trade_price * (1 - self.params.profit_goal_neg)
                self.entry_index = len(self)
                self.short_trade = True

class PandasData(bt.feeds.PandasData):
    lines = ('hourly_returns',)
    params = (('hourly_returns', 6),)  

def objective_short(trial, df):
    k_pos = trial.suggest_float('k_pos', 1.5, 2.25)
    window_pos = trial.suggest_int('window_pos', 50, 600)
    n_pos = trial.suggest_int('n_pos', 8, 30)
    stop_loss_pos = trial.suggest_float('stop_loss_pos', 0.025, 0.075)
    lookback_pos = trial.suggest_int('lookback_pos', 3, 12)
    profit_goal_pos = trial.suggest_float('profit_goal_pos', 0.025, 0.1)
    
    k_neg = trial.suggest_float('k_neg', 1.5, 2.3)
    window_neg = trial.suggest_int('window_neg', 40, 600)
    n_neg = trial.suggest_int('n_neg', 10, 30)
    stop_loss_neg = trial.suggest_float('stop_loss_neg', 0.025, 0.075)
    lookback_neg = trial.suggest_int('lookback_neg', 3, 12)
    profit_goal_neg = trial.suggest_float('profit_goal_neg', 0.02, 0.1)

    cerebro = bt.Cerebro()
    cerebro.addstrategy(CombinedShortingStrategy, 
                        k_pos=k_pos, window_pos=window_pos, n_pos=n_pos, stop_loss_pos=stop_loss_pos, lookback_pos=lookback_pos, profit_goal_pos=profit_goal_pos,
                        k_neg=k_neg, window_neg=window_neg, n_neg=n_neg, stop_loss_neg=stop_loss_neg, lookback_neg=lookback_neg, profit_goal_neg=profit_goal_neg)
    
    data = PandasData(
        dataname=df,
        datetime='date',
        open='open',
        high='high',
        low='low',
        close='close',
        volume='volume',
        openinterest=-1,
        hourly_returns='hourly_returns'
    )
    cerebro.adddata(data)
    
    cerebro.broker.set_cash(100000)#common starting value
    cerebro.run()
    portfolio_value = cerebro.broker.get_value()
    
    if np.isnan(portfolio_value):
        return -np.inf  # instead of nan
    
    return portfolio_value
```

### Parameters Search

The parameters for the strategies will be found using optuna search to save time and archieve max efficiency. The list of hyperparameters are decided based on bias and limitations of the research to reduce overfitting and look-ahead biases. It is based on the maximum profit per trade to find the best parameters to reduce useless trades and maximize profits which transaction costs is thought of.

The use of training data set is to find the appropriate parameters for the testing set. 

There will be a set of parameters for both postive and negative returns/trades.

- k (1.5-2.25) the number of sd for the threshold, (cumalitive returns over rolling mean and k rolling sd)
- window (50-600) window for moving average and rolling sd
- n (8-30) number of hours staying in a trade if didn't exit through stop loss or profit goal
- stop loss (0.025-0.075) exit signal 
- lookback (3-12) number of hours looking back for the cumalitive returns
- profit goal (0.025-0.1) exit signal

This means there are 12 parameters per coin per strategy, and 24 parameters for the volatility clustering.

```python
class PerformanceAnalyzer(bt.Analyzer): #for perfermance metrics
    def create_analysis(self):
        self.rets = []

    def notify_trade(self, trade):
        if trade.isclosed:
            self.rets.append(trade.pnlcomm / trade.price)  

    def get_analysis(self):
        return self.rets

def run_long_strategy_on_test_set(test_df, params):
    cerebro = bt.Cerebro()
    cerebro.addstrategy(CombinedStrategy,
                        k_pos=params['k_pos'], window_pos=params['window_pos'], n_pos=params['n_pos'], stop_loss_pos=params['stop_loss_pos'],
                        lookback_pos=params['lookback_pos'], profit_goal_pos=params['profit_goal_pos'],
                        k_neg=params['k_neg'], window_neg=params['window_neg'], n_neg=params['n_neg'], stop_loss_neg=params['stop_loss_neg'],
                        lookback_neg=params['lookback_neg'], profit_goal_neg=params['profit_goal_neg'],
                        transaction_cost=0.0025)
    
    data = PandasData(
        dataname=test_df,
        datetime='date',
        open='open',
        high='high',
        low='low',
        close='close',
        volume='volume',
        openinterest=-1,
        hourly_returns='hourly_returns'
    )
    cerebro.adddata(data)
    cerebro.broker.set_cash(100000) #common starting value
    cerebro.addanalyzer(bt.analyzers.PyFolio, _name='pyfolio')
    cerebro.addanalyzer(bt.analyzers.SharpeRatio, _name='sharpe')
    cerebro.addanalyzer(bt.analyzers.DrawDown, _name='drawdown')
    cerebro.addanalyzer(PerformanceAnalyzer, _name='performance')
    
    results = cerebro.run()
    strat = results[0]
    
    sharpe_ratio = strat.analyzers.sharpe.get_analysis()['sharperatio']
    max_drawdown = strat.analyzers.drawdown.get_analysis()['max']['drawdown']
    
    returns = strat.analyzers.performance.get_analysis()
    total_profit = np.sum(returns) * 100  # percentage
    num_trades = len(returns)
    profit_per_trade = total_profit / num_trades if num_trades > 0 else 0
    
    return {
        'sharpe_ratio': sharpe_ratio,
        'max_drawdown': max_drawdown,
        'returns_summary': {
            'mean': np.mean(returns),
            'std': np.std(returns),
            'min': np.min(returns),
            'max': np.max(returns),
            'total': total_profit,
            'count': num_trades
        },
        'total_profit': total_profit,
        'profit_per_trade': profit_per_trade
    }

def run_short_strategy_on_test_set(test_df, params):
    cerebro = bt.Cerebro()
    cerebro.addstrategy(CombinedShortingStrategy,
                        k_pos=params['k_pos'], window_pos=params['window_pos'], n_pos=params['n_pos'], stop_loss_pos=params['stop_loss_pos'],
                        lookback_pos=params['lookback_pos'], profit_goal_pos=params['profit_goal_pos'],
                        k_neg=params['k_neg'], window_neg=params['window_neg'], n_neg=params['n_neg'], stop_loss_neg=params['stop_loss_neg'],
                        lookback_neg=params['lookback_neg'], profit_goal_neg=params['profit_goal_neg'],
                        transaction_cost=0.0025)
    
    data = PandasData(
        dataname=test_df,
        datetime='date',
        open='open',
        high='high',
        low='low',
        close='close',
        volume='volume',
        openinterest=-1,
        hourly_returns='hourly_returns'
    )
    cerebro.adddata(data)
    
    cerebro.broker.set_cash(100000)
    cerebro.addanalyzer(bt.analyzers.PyFolio, _name='pyfolio')
    cerebro.addanalyzer(bt.analyzers.SharpeRatio, _name='sharpe')
    cerebro.addanalyzer(bt.analyzers.DrawDown, _name='drawdown')
    cerebro.addanalyzer(PerformanceAnalyzer, _name='performance')
    
    results = cerebro.run()
    strat = results[0]
    
    sharpe_ratio = strat.analyzers.sharpe.get_analysis()['sharperatio']
    max_drawdown = strat.analyzers.drawdown.get_analysis()['max']['drawdown']
    
    returns = strat.analyzers.performance.get_analysis()
    total_profit = np.sum(returns) * 100  # percentage
    num_trades = len(returns)
    profit_per_trade = total_profit / num_trades if num_trades > 0 else 0
    
    return {
        'sharpe_ratio': sharpe_ratio,
        'max_drawdown': max_drawdown,
        'returns_summary': {
            'mean': np.mean(returns),
            'std': np.std(returns),
            'min': np.min(returns),
            'max': np.max(returns),
            'total': total_profit,
            'count': num_trades
        },
        'total_profit': total_profit,
        'profit_per_trade': profit_per_trade
    }
```

## Walk forward analysis

To apply walk forward analysis, we split the the test period and training period into overlapping periods and fine-tune the parameters as we look at each new test set.

```python
def walk_forward_analysis_long(df, initial_train_period, test_period, n_trials):
    results = []
    start_date = df['date'].min()
    end_date = df['date'].max()
    period_start = start_date

    while period_start + initial_train_period + test_period <= end_date:
        train_end = period_start + initial_train_period
        test_end = train_end + test_period

        train_df = df[(df['date'] >= period_start) & (df['date'] < train_end)]
        test_df = df[(df['date'] >= train_end) & (df['date'] < test_end)]
        study = optuna.create_study(direction='maximize')
        study.optimize(lambda trial: objective_long(trial, train_df), n_trials=n_trials, n_jobs=-1)
        best_params = study.best_params

        test_portfolio_value = run_long_strategy_on_test_set(test_df, best_params)
        results.append({
            'train_period_start': period_start,
            'train_period_end': train_end,
            'test_period_start': train_end,
            'test_period_end': test_end,
            'test_portfolio_value': test_portfolio_value,
            'best_params': best_params
        })

        period_start += test_period  #move to the next period for walk forward

    return pd.DataFrame(results)

def walk_forward_analysis_short(df, initial_train_period, test_period, n_trials):
    results = []
    start_date = df['date'].min()
    end_date = df['date'].max()
    period_start = start_date

    while period_start + initial_train_period + test_period <= end_date:
        train_end = period_start + initial_train_period
        test_end = train_end + test_period

        train_df = df[(df['date'] >= period_start) & (df['date'] < train_end)]
        test_df = df[(df['date'] >= train_end) & (df['date'] < test_end)]
        study = optuna.create_study(direction='maximize')
        study.optimize(lambda trial: objective_short(trial, train_df), n_trials=n_trials, n_jobs=-1)
        best_params = study.best_params

        test_portfolio_value = run_short_strategy_on_test_set(test_df, best_params)
        results.append({
            'train_period_start': period_start,
            'train_period_end': train_end,
            'test_period_start': train_end,
            'test_period_end': test_end,
            'test_portfolio_value': test_portfolio_value,
            'best_params': best_params
        })

        period_start += test_period  #move to the next period for walk forward

    return pd.DataFrame(results)
```

## Volatility Clustering


Here is the second part of the strategy where we look at the classification of two different volatility phases to have different parameters for each phase for the strategy. First we will be looking at the distributions and graphs for sanity check. The motive behind the volatility clustering is the emprical evidence from two papers: Intraday downward/upward multifractality and long memory in Bitcoin and Ethereum markets: An asymmetric multifractal detrended fluctuation analysis.(Mensi, 2019) and MAX momentum in cryptocurrency markets.(Wang, P., & Zhang, W. 2021). Two papers both found that the returns of BTC are different over time and evidence of volatility clustering. That's why I decided to take a closer look at the volatility clustering in this strategy which will be an extension to the hourly momentum strategy.



### ATR

First method of classifying the volatility is the Average True Range (ATR), used to measure market volatility. It provides insight into how much an asset's price is moving on average over a specified period. 

The True Range (TR) for a given period \( t \) is defined as the maximum of the following:
\begin{equation}
TR_t = \max(H_t - L_t, |H_t - C_{t-1}|, |L_t - C_{t-1}|)
\end{equation}
where \( H_t \) is the high price, \( L_t \) is the low price, and \( C_{t-1} \) is the previous close price.


The ATR over \( n \) periods is calculated as the rolling mean of the True Range values:
\begin{equation}
ATR_n = \frac{1}{n} \sum_{i=0}^{n-1} TR_{t-i}
\end{equation}

### GARCH

The second method we use will be the Generalized Autoregressive Conditional Heteroskedasticity (GARCH) model. It is used more specifically to estimate the volatility of financial returns and timeseries data which might suit our case. 

A GARCH(1,1) model is:
\begin{equation}
\sigma_t^2 = \alpha_0 + \alpha_1 \epsilon_{t-1}^2 + \beta_1 \sigma_{t-1}^2
\end{equation}
where:
$\sigma_t^2$ is the conditional variance at time $t$, $\alpha_0$ is the constant term,
$\alpha_1$ is the coefficient for the lagged squared residual ($\epsilon_{t-1}^2$),
 and $\beta_1$ is the coefficient for the lagged conditional variance ($\sigma_{t-1}^2$).


Similar to the ATR model, we will be identifying high volatility periods using the GARCH model, the trading strategy will then adapt to the two different market conditions by applying different parameter sets for high and normal volatility phases.

## Volatility clustering momentum/Contrarian Strategy rules


This will be the second part of the strategy where incorporate the volatility clusterings on top of the existing strategy. The rules will be the same, but the parameters will all be separate for the high volatility phases.

### ATR-Based 


1. Calculate the True Range (TR): 

   $$
   TR_t = \max(H_t - L_t, |H_t - C_{t-1}|, |L_t - C_{t-1}|)
   $$
   
   where $ H_t $ is the high price, $ L_t $ is the low price, and $ C_{t-1} $ is the previous close price.


2. Compute the ATR:

   $$
   ATR_n = \frac{1}{n} \sum_{i=0}^{n-1} TR_{t-i}
   $$


3. Define the Volatility Threshold: A threshold $ \theta $ is chosen to classify high and low volatility period from the T-test above:

   $$
   \theta = \text{percentile}(ATR, 90)
   $$


4. Classify Volatility Phases: Using the ATR and the threshold $\theta$, classify each period as high or normal volatility:
$$
\text{high volatility}_t = 
\begin{cases} 
1 & \text{if } ATR_t > \theta \\
0 & \text{if } ATR_t \leq \theta 
\end{cases}
$$





After identifying the phases, there will be momentum/contrarian strategy for both phases but disregarding the momentum side for negative abnormal returns



Mean and Standard Deviation of the rolling window for hours:(Indicator)
$$
\mu_t = \frac{1}{n} \sum_{i=t-n+1}^{t} R_i
$$

$$
\sigma_t = \sqrt{\frac{1}{n-1} \sum_{i=t-n+1}^{t} (R_i - \mu_t)^2}
$$

Calculation of the cumulative return in the look back window: (Indicator)
$$
C_t = \sum_{i=t-\text{lookback}+1}^{t} R_i
$$

Threshold for postive and negative returns:(Indicator)
$$
\text{Threshold}_t = \mu_t + k \cdot \sigma_t \cdot \sqrt{\text{lookback}}
$$
$$
\text{Threshold}_t = \mu_t - k \cdot \sigma_t \cdot \sqrt{\text{lookback}}
$$

Enter trade for postive and negative returns:(signal)
$$
C_t > \text{Threshold}_t
$$
$$
C_t < \text{Threshold}_t
$$


The trade is exited after holding for n periods or reached profit goal or if the stop loss condition is met:


Stop loss: (signal) postive momentum trading and negative contrarian trading would take long postions.
$$
\text{Stop Loss Condition for long position: } \frac{P_{t+j} - P_t}{P_t} \leq -\text{stop loss}
$$




```python
class CombinedStrategyVol(bt.Strategy): #no more short strategy since proven long for negative return is better
    params = (
        #high volatility
        ('k_pos_high', 1.5),
        ('window_pos_high', 50),
        ('n_pos_high', 5),
        ('stop_loss_pos_high', 0.05),
        ('lookback_pos_high', 3),
        ('profit_goal_pos_high', 0.025),
        ('k_neg_high', 1.5),
        ('window_neg_high', 50),
        ('n_neg_high', 5),
        ('stop_loss_neg_high', 0.05),
        ('lookback_neg_high', 3),
        ('profit_goal_neg_high', 0.025),
        #low volatility
        ('k_pos_low', 1.5),
        ('window_pos_low', 50),
        ('n_pos_low', 5),
        ('stop_loss_pos_low', 0.05),
        ('lookback_pos_low', 3),
        ('profit_goal_pos_low', 0.025),
        ('k_neg_low', 1.5),
        ('window_neg_low', 50),
        ('n_neg_low', 5),
        ('stop_loss_neg_low', 0.05),
        ('lookback_neg_low', 3),
        ('profit_goal_neg_low', 0.025),
        ('transaction_cost', 0.0025),  # 0.25%
    )

    def __init__(self):
        self.dataclose = self.data.close
        self.pct_change = self.data.hourly_returns
        self.high_volatility = self.data.high_volatility

        #indicators for high volatility 
        self.rolling_mean_pos_high = bt.indicators.SimpleMovingAverage(self.pct_change, period=self.params.window_pos_high)
        self.rolling_std_pos_high = bt.indicators.StandardDeviation(self.pct_change, period=self.params.window_pos_high) * np.sqrt(self.params.lookback_pos_high)
        self.cumulative_returns_pos_high = bt.indicators.SumN(self.pct_change, period=self.params.lookback_pos_high)
        self.positive_threshold_high = self.rolling_mean_pos_high + self.params.k_pos_high * self.rolling_std_pos_high

        self.rolling_mean_neg_high = bt.indicators.SimpleMovingAverage(self.pct_change, period=self.params.window_neg_high)
        self.rolling_std_neg_high = bt.indicators.StandardDeviation(self.pct_change, period=self.params.window_neg_high) * np.sqrt(self.params.lookback_neg_high)
        self.cumulative_returns_neg_high = bt.indicators.SumN(self.pct_change, period=self.params.lookback_neg_high)
        self.negative_threshold_high = self.rolling_mean_neg_high - self.params.k_neg_high * self.rolling_std_neg_high

        #indicators for low volatility 
        self.rolling_mean_pos_low = bt.indicators.SimpleMovingAverage(self.pct_change, period=self.params.window_pos_low)
        self.rolling_std_pos_low = bt.indicators.StandardDeviation(self.pct_change, period=self.params.window_pos_low) * np.sqrt(self.params.lookback_pos_low)
        self.cumulative_returns_pos_low = bt.indicators.SumN(self.pct_change, period=self.params.lookback_pos_low)
        self.positive_threshold_low = self.rolling_mean_pos_low + self.params.k_pos_low * self.rolling_std_pos_low

        self.rolling_mean_neg_low = bt.indicators.SimpleMovingAverage(self.pct_change, period=self.params.window_neg_low)
        self.rolling_std_neg_low = bt.indicators.StandardDeviation(self.pct_change, period=self.params.window_neg_low) * np.sqrt(self.params.lookback_neg_low)
        self.cumulative_returns_neg_low = bt.indicators.SumN(self.pct_change, period=self.params.lookback_neg_low)
        self.negative_threshold_low = self.rolling_mean_neg_low - self.params.k_neg_low * self.rolling_std_neg_low

        self.in_trade = False
        self.trade_price = None
        self.stop_loss_price = None
        self.profit_goal_price = None
        self.entry_index = None

    def next(self):
        if self.in_trade:
            # Exit conditions
            if self.dataclose[0] <= self.stop_loss_price or self.dataclose[0] >= self.profit_goal_price or len(self) - self.entry_index >= self.params.n_pos_high:
                self.sell()
                self.in_trade = False

        if not self.in_trade:
            if self.high_volatility[0]:
                #positive momentum, high volatility
                if self.cumulative_returns_pos_high[0] > self.positive_threshold_high[0]:
                    self.buy()
                    self.in_trade = True
                    self.trade_price = self.dataclose[0] * (1 + self.params.transaction_cost)  # transaction cost
                    self.stop_loss_price = self.trade_price * (1 - self.params.stop_loss_pos_high)
                    self.profit_goal_price = self.trade_price * (1 + self.params.profit_goal_pos_high)
                    self.entry_index = len(self)
                #negative contrarian (long) strategy for high volatility
                elif self.cumulative_returns_neg_high[0] < self.negative_threshold_high[0]:
                    self.buy()
                    self.in_trade = True
                    self.trade_price = self.dataclose[0] * (1 + self.params.transaction_cost)  # transaction cost
                    self.stop_loss_price = self.trade_price * (1 - self.params.stop_loss_neg_high)
                    self.profit_goal_price = self.trade_price * (1 + self.params.profit_goal_neg_high)
                    self.entry_index = len(self)
            else:
                if self.cumulative_returns_pos_low[0] > self.positive_threshold_low[0]:
                    self.buy()
                    self.in_trade = True
                    self.trade_price = self.dataclose[0] * (1 + self.params.transaction_cost)  # transaction cost
                    self.stop_loss_price = self.trade_price * (1 - self.params.stop_loss_pos_low)
                    self.profit_goal_price = self.trade_price * (1 + self.params.profit_goal_pos_low)
                    self.entry_index = len(self)
                elif self.cumulative_returns_neg_low[0] < self.negative_threshold_low[0]:
                    self.buy()
                    self.in_trade = True
                    self.trade_price = self.dataclose[0] * (1 + self.params.transaction_cost)  # transaction cost
                    self.stop_loss_price = self.trade_price * (1 - self.params.stop_loss_neg_low)
                    self.profit_goal_price = self.trade_price * (1 + self.params.profit_goal_neg_low)
                    self.entry_index = len(self)
                    
def objective_long_vol(trial, df):
    k_pos_high = trial.suggest_float('k_pos_high', 1.5, 2.5)
    window_pos_high = trial.suggest_int('window_pos_high', 20, 100)
    n_pos_high = trial.suggest_int('n_pos_high', 1, 10)
    stop_loss_pos_high = trial.suggest_float('stop_loss_pos_high', 0.01, 0.1)
    lookback_pos_high = trial.suggest_int('lookback_pos_high', 1, 10)
    profit_goal_pos_high = trial.suggest_float('profit_goal_pos_high', 0.01, 0.1)

    k_neg_high = trial.suggest_float('k_neg_high', 1.5, 2.5)
    window_neg_high = trial.suggest_int('window_neg_high', 20, 100)
    n_neg_high = trial.suggest_int('n_neg_high', 1, 10)
    stop_loss_neg_high = trial.suggest_float('stop_loss_neg_high', 0.01, 0.1)
    lookback_neg_high = trial.suggest_int('lookback_neg_high', 1, 10)
    profit_goal_neg_high = trial.suggest_float('profit_goal_neg_high', 0.01, 0.1)

    k_pos_low = trial.suggest_float('k_pos_low', 1.5, 2.5)
    window_pos_low = trial.suggest_int('window_pos_low', 20, 100)
    n_pos_low = trial.suggest_int('n_pos_low', 1, 10)
    stop_loss_pos_low = trial.suggest_float('stop_loss_pos_low', 0.01, 0.1)
    lookback_pos_low = trial.suggest_int('lookback_pos_low', 1, 10)
    profit_goal_pos_low = trial.suggest_float('profit_goal_pos_low', 0.01, 0.1)

    k_neg_low = trial.suggest_float('k_neg_low', 1.5, 2.5)
    window_neg_low = trial.suggest_int('window_neg_low', 20, 100)
    n_neg_low = trial.suggest_int('n_neg_low', 1, 10)
    stop_loss_neg_low = trial.suggest_float('stop_loss_neg_low', 0.01, 0.1)
    lookback_neg_low = trial.suggest_int('lookback_neg_low', 1, 10)
    profit_goal_neg_low = trial.suggest_float('profit_goal_neg_low', 0.01, 0.1)

    cerebro = bt.Cerebro()
    cerebro.addstrategy(CombinedStrategyVol,
                        k_pos_high=k_pos_high, window_pos_high=window_pos_high, n_pos_high=n_pos_high, stop_loss_pos_high=stop_loss_pos_high,
                        lookback_pos_high=lookback_pos_high, profit_goal_pos_high=profit_goal_pos_high,
                        k_neg_high=k_neg_high, window_neg_high=window_neg_high, n_neg_high=n_neg_high, stop_loss_neg_high=stop_loss_neg_high,
                        lookback_neg_high=lookback_neg_high, profit_goal_neg_high=profit_goal_neg_high,
                        k_pos_low=k_pos_low, window_pos_low=window_pos_low, n_pos_low=n_pos_low, stop_loss_pos_low=stop_loss_pos_low,
                        lookback_pos_low=lookback_pos_low, profit_goal_pos_low=profit_goal_pos_low,
                        k_neg_low=k_neg_low, window_neg_low=window_neg_low, n_neg_low=n_neg_low, stop_loss_neg_low=stop_loss_neg_low,
                        lookback_neg_low=lookback_neg_low, profit_goal_neg_low=profit_goal_neg_low,
                        transaction_cost=0.0025)

    data = PandasDataVol(
        dataname=df,
        datetime='date',
        open='open',
        high='high',
        low='low',
        close='close',
        volume='volume',
        openinterest=-1,
        hourly_returns='hourly_returns',
        high_volatility='high_volatility'
    )
    cerebro.adddata(data)
    cerebro.broker.set_cash(100000)
    cerebro.run()

    portfolio_value = cerebro.broker.get_value()

    if np.isnan(portfolio_value):
        return -np.inf  # nan solve

    return portfolio_value
class PandasDataVol(bt.feeds.PandasData):
    lines = ('hourly_returns', 'high_volatility')
    params = (('hourly_returns', 6), ('high_volatility', 7))

#Function to calculate ATR and classify volatility
def calculate_atr_vol(df, atr_period=12, atr_threshold=0.75):
    df['high_low'] = df['high'] - df['low']
    df['high_close'] = np.abs(df['high'] - df['close'].shift(1))
    df['low_close'] = np.abs(df['low'] - df['close'].shift(1))
    df['true_range'] = df[['high_low', 'high_close', 'low_close']].max(axis=1)
    
    atr_col = f'atr_{atr_period}'
    df[atr_col] = df['true_range'].rolling(window=atr_period).mean()
    
    threshold = df[atr_col].quantile(atr_threshold)
    df['high_volatility'] = df[atr_col] > threshold
    
    df.drop(['high_low', 'high_close', 'low_close'], axis=1, inplace=True)
    df.dropna(subset=[atr_col], inplace=True)
    
    return df

def run_long_strategy_on_test_set_vol(test_df, params, period, threshold):
    test_df = calculate_atr_vol(test_df, period, threshold)

    cerebro = bt.Cerebro()
    cerebro.addstrategy(CombinedStrategyVol,
                        k_pos_high=params['k_pos_high'], window_pos_high=params['window_pos_high'], n_pos_high=params['n_pos_high'], stop_loss_pos_high=params['stop_loss_pos_high'],
                        lookback_pos_high=params['lookback_pos_high'], profit_goal_pos_high=params['profit_goal_pos_high'],
                        k_neg_high=params['k_neg_high'], window_neg_high=params['window_neg_high'], n_neg_high=params['n_neg_high'], stop_loss_neg_high=params['stop_loss_neg_high'],
                        lookback_neg_high=params['lookback_neg_high'], profit_goal_neg_high=params['profit_goal_neg_high'],
                        k_pos_low=params['k_pos_low'], window_pos_low=params['window_pos_low'], n_pos_low=params['n_pos_low'], stop_loss_pos_low=params['stop_loss_pos_low'],
                        lookback_pos_low=params['lookback_pos_low'], profit_goal_pos_low=params['profit_goal_pos_low'],
                        k_neg_low=params['k_neg_low'], window_neg_low=params['window_neg_low'], n_neg_low=params['n_neg_low'], stop_loss_neg_low=params['stop_loss_neg_low'],
                        lookback_neg_low=params['lookback_neg_low'], profit_goal_neg_low=params['profit_goal_neg_low'],
                        transaction_cost=0.0025)
    
    data = PandasDataVol(
        dataname=test_df,
        datetime='date',
        open='open',
        high='high',
        low='low',
        close='close',
        volume='volume',
        openinterest=-1,
        hourly_returns='hourly_returns',
        high_volatility='high_volatility'
    )
    cerebro.adddata(data)
    
    cerebro.broker.set_cash(100000)
    cerebro.addanalyzer(bt.analyzers.PyFolio, _name='pyfolio')
    cerebro.addanalyzer(bt.analyzers.SharpeRatio, _name='sharpe')
    cerebro.addanalyzer(bt.analyzers.DrawDown, _name='drawdown')
    
    results = cerebro.run()
    strat = results[0]
    
    returns, positions, transactions, gross_lev = strat.analyzers.pyfolio.get_pf_items()
    sharpe_ratio = strat.analyzers.sharpe.get_analysis().get('sharperatio')
    max_drawdown = strat.analyzers.drawdown.get_analysis().get('max', {}).get('drawdown', np.nan)
    
    returns_summary = {
        'mean': np.mean(returns),
        'std': np.std(returns),
        'min': np.min(returns),
        'max': np.max(returns),
        'total': np.sum(returns),
        'count': len(returns)
    }
    
    total_profit = np.sum(returns) * 100  # percentage
    profit_per_trade = total_profit / len(returns) if len(returns) > 0 else 0
    
    return {
        'sharpe_ratio': sharpe_ratio,
        'max_drawdown': max_drawdown,
        'returns_summary': returns_summary,
        'total_profit': total_profit,
        'profit_per_trade': profit_per_trade
    }

def optimize_and_test_vol(train_df, test_df, strategy_type, period=12, threshold=0.75):
    train_df = calculate_atr_vol(train_df, period, threshold)
    test_df = calculate_atr_vol(test_df, period, threshold)

    study = optuna.create_study(direction='maximize')
    study.optimize(lambda trial: objective_long_vol(trial, train_df), n_trials=20, n_jobs=-1)

    best_params = study.best_params

    test_results = run_long_strategy_on_test_set_vol(test_df, best_params, period, threshold)

    return test_results, best_params
```

```python
def walk_forward_analysis_long_vol(df, initial_train_period, test_period, n_trials, period, threshold):
    results = []
    start_date = df['date'].min()
    end_date = df['date'].max()
    period_start = start_date

    while period_start + initial_train_period + test_period <= end_date:
        train_end = period_start + initial_train_period
        test_end = train_end + test_period

        train_df = df[(df['date'] >= period_start) & (df['date'] < train_end)]
        test_df = df[(df['date'] >= train_end) & (df['date'] < test_end)]

        train_df = calculate_atr_vol(train_df, period, threshold)

        study = optuna.create_study(direction='maximize')
        study.optimize(lambda trial: objective_long_vol(trial, train_df), n_trials=n_trials, n_jobs=-1)
        best_params = study.best_params

        test_df = calculate_atr_vol(test_df, period, threshold)

        test_portfolio_value = run_long_strategy_on_test_set_vol(test_df, best_params, period, threshold)
        results.append({
            'train_period_start': period_start,
            'train_period_end': train_end,
            'test_period_start': train_end,
            'test_period_end': test_end,
            'test_portfolio_value': test_portfolio_value,
            'best_params': best_params
        })

        period_start += test_period  

    return pd.DataFrame(results)
```



### GARCH-Based 



1. Fit the GARCH Model: 
   $$
   \sigma_t^2 = \omega + \alpha \epsilon_{t-1}^2 + \beta \sigma_{t-1}^2
   $$
   where $\sigma_t^2 $ is the conditional variance at time $t $, $ \omega $ is a constant, $ \alpha $ is the coefficient for the lagged squared residual, and $ \beta $ is the coefficient for the lagged variance.

2. Estimate Conditional Volatility $\sigma_t $ for each period $ t $:

3. Define the Volatility Threshold: Similar to the ATR method, define a threshold $ \theta $ based on the distribution of estimated volatilities from signal testing above:
   $$
   \theta = \text{percentile}(\sigma, 90)
   $$

4. Classify Volatility Phases: Using the conditional volatility estimates and the threshold $ \theta $, classify each period as high or normal volatility:
   $$
   \text{high volatility}_t = 
   \begin{cases} 
   1 & \text{if } \sigma_t > \theta \\
   0 & \text{if } \sigma_t \leq \theta 
   \end{cases}
   $$

After identifying the phases, there will be momentum/contrarian strategy for both phases but disregarding the momentum side for negative abnormal returns



Mean and Standard Deviation of the rolling window for hours:(Indicator)
$$
\mu_t = \frac{1}{n} \sum_{i=t-n+1}^{t} R_i
$$

$$
\sigma_t = \sqrt{\frac{1}{n-1} \sum_{i=t-n+1}^{t} (R_i - \mu_t)^2}
$$

Calculation of the cumulative return in the look back window: (Indicator)
$$
C_t = \sum_{i=t-\text{lookback}+1}^{t} R_i
$$

Threshold for postive and negative returns:(Indicator)
$$
\text{Threshold}_t = \mu_t + k \cdot \sigma_t \cdot \sqrt{\text{lookback}}
$$
$$
\text{Threshold}_t = \mu_t - k \cdot \sigma_t \cdot \sqrt{\text{lookback}}
$$

Enter trade for postive and negative returns:(signal)
$$
C_t > \text{Threshold}_t
$$
$$
C_t < \text{Threshold}_t
$$


The trade is exited after holding for n periods or reached profit goal or if the stop loss condition is met:


Stop loss: (signal) postive momentum trading and negative contrarian trading would take long postions.
$$
\text{Stop Loss Condition for long position: } \frac{P_{t+j} - P_t}{P_t} \leq -\text{stop loss}

```python
def walk_forward_analysis_long_vol2(df, initial_train_period, test_period, n_trials, threshold):
    results = []
    start_date = df['date'].min()
    end_date = df['date'].max()
    period_start = start_date

    while period_start + initial_train_period + test_period <= end_date:
        train_end = period_start + initial_train_period
        test_end = train_end + test_period

        train_df = df[(df['date'] >= period_start) & (df['date'] < train_end)]
        test_df = df[(df['date'] >= train_end) & (df['date'] < test_end)]

        train_df = calculate_garch_vol(train_df, threshold)

        study = optuna.create_study(direction='maximize')
        study.optimize(lambda trial: objective_long_vol(trial, train_df), n_trials=n_trials, n_jobs=-1)
        best_params = study.best_params

        test_df = calculate_garch_vol(test_df, threshold)

        test_portfolio_value = run_long_strategy_on_test_set_vol2(test_df, best_params, threshold)
        results.append({
            'train_period_start': period_start,
            'train_period_end': train_end,
            'test_period_start': train_end,
            'test_period_end': test_end,
            'test_portfolio_value': test_portfolio_value,
            'best_params': best_params
        })

        period_start += test_period  

    return pd.DataFrame(results)
```

```python
def calculate_holding_returns(df): #holding
    start_price = df['close'].iloc[0]
    end_price = df['close'].iloc[-1]
    returns = (df['close'] / df['close'].shift(1)).dropna() - 1
    holding_return = (end_price - start_price) / start_price
    return returns, holding_return


def random_trading_simulator(df, n_trades=100): # random trading
    cash = 100000
    position = 0  
    trade_size = cash / n_trades
    trade_dates = random.sample(range(len(df)), n_trades)
    
    returns = []
    for date in trade_dates:
        trade_action = random.choice(['buy', 'sell', 'hold'])
        price = df['close'].iloc[date]
        if trade_action == 'buy' and cash >= trade_size:
            cash -= trade_size
            position += trade_size / price
        elif trade_action == 'sell' and position > 0:
            cash += position * price
            position = 0

        daily_return = (price / df['close'].iloc[date - 1]) - 1 if date > 0 else 0
        returns.append(daily_return)

    final_price = df['close'].iloc[-1]
    final_value = cash + position * final_price
    total_return = (final_value - 100000) / 100000
    return pd.Series(returns), total_return


def cash_simulator(): #  cash 
    return pd.Series([0]), 0

def calculate_performance_metrics(returns):
    sharpe_ratio = returns.mean() / returns.std() * np.sqrt(252) if returns.std() != 0 else np.nan
    cumulative_returns = (1 + returns).cumprod()
    max_drawdown = (cumulative_returns.cummax() - cumulative_returns).max()
    mean_return = returns.mean()
    std_return = returns.std()
    min_return = returns.min()
    max_return = returns.max()
    total_return = returns.sum()
    trade_count = len(returns)
    profit_per_trade = total_return / trade_count if trade_count > 0 else 0

    return {
        'Sharpe Ratio': sharpe_ratio,
        'Max Drawdown': max_drawdown,
        'Mean Return': mean_return,
        'Std Dev Return': std_return,
        'Min Return': min_return,
        'Max Return': max_return,
        'Total Return': total_return,
        'Trade Count': trade_count,
        'Profit per Trade': profit_per_trade
    }

def calculate_benchmarks(btc_df, eth_df, sol_df):
    btc_holding_returns, btc_holding_total = calculate_holding_returns(btc_df)
    eth_holding_returns, eth_holding_total = calculate_holding_returns(eth_df)
    sol_holding_returns, sol_holding_total = calculate_holding_returns(sol_df)

    btc_random_returns, btc_random_total = random_trading_simulator(btc_df)
    eth_random_returns, eth_random_total = random_trading_simulator(eth_df)
    sol_random_returns, sol_random_total = random_trading_simulator(sol_df)

    cash_returns, cash_total = cash_simulator()

    holding_metrics = {
        'BTC Holding': calculate_performance_metrics(btc_holding_returns),
        'ETH Holding': calculate_performance_metrics(eth_holding_returns),
        'SOL Holding': calculate_performance_metrics(sol_holding_returns),
    }

    random_trading_metrics = {
        'BTC Random Trading': calculate_performance_metrics(btc_random_returns),
        'ETH Random Trading': calculate_performance_metrics(eth_random_returns),
        'SOL Random Trading': calculate_performance_metrics(sol_random_returns),
    }

    cash_metrics = {
        'Cash': calculate_performance_metrics(cash_returns),
    }

    return {**holding_metrics, **random_trading_metrics, **cash_metrics}

btc_benchmarks = calculate_benchmarks(test_btc_df, test_eth_df, test_sol_df)

benchmarks_rows = []
for strategy, metrics in btc_benchmarks.items():
    row = {
        'Strategy': strategy,
        'Sharpe Ratio': metrics['Sharpe Ratio'],
        'Max Drawdown': metrics['Max Drawdown'],
        'Mean Return': metrics['Mean Return'],
        'Std Dev Return': metrics['Std Dev Return'],
        'Min Return': metrics['Min Return'],
        'Max Return': metrics['Max Return'],
        'Total Return': metrics['Total Return'],
        'Trade Count': metrics['Trade Count'],
        'Profit per Trade': metrics['Profit per Trade']
    }
    benchmarks_rows.append(row)

benchmarks_df = pd.DataFrame(benchmarks_rows)
benchmarks_df.set_index('Strategy', inplace=True)
#Combine all results
combined_results_df = pd.concat([results_df1,results_df_vol,results_df_garch, benchmarks_df])
```

```python
def strategy_evaluation_positive(hourly_df, k=1.5, window=50, n=5, stop_loss=0.05, lookback=3, profit_goal=0.025,name='BTCUSD'):
    hourly_df = hourly_df.copy()
    hourly_df['hourly_returns'] = hourly_df['close'].pct_change()
    hourly_df.dropna(inplace=True)
    hourly_df['rolling_mean'] = hourly_df['hourly_returns'].rolling(window=window).mean() #rolling mean
    hourly_df['rolling_std'] = hourly_df['hourly_returns'].rolling(window=window).std() * np.sqrt(lookback) #depends on lookback

    #cumalitive returns
    hourly_df['cumulative_returns'] = hourly_df['hourly_returns'].rolling(window=lookback).sum()
    hourly_df['threshold'] = hourly_df['rolling_mean'] + k * hourly_df['rolling_std'] #threshold
    
    results = []
    profits = []
    in_trade = False  # THis shows that we are in trade or not so not overlap trade

    for i in range(len(hourly_df) - n):
        if in_trade:
            continue

        if hourly_df.iloc[i]['cumulative_returns'] > hourly_df.iloc[i]['threshold']:  
            opening_price = hourly_df.iloc[i]['close']
            closing_price = opening_price
            stop_hit = False

            # Enter trade
            in_trade = True

            for j in range(1, n + 1):
                current_price = hourly_df.iloc[i + j]['close']
                if (current_price - opening_price) / opening_price <= -stop_loss:  #end trade for stoploss
                    closing_price = current_price
                    stop_hit = True
                    break
                if (current_price - opening_price) / opening_price >= profit_goal:  #profit goal condition
                    closing_price = current_price
                    break
                closing_price = current_price #else end trade after n
            
            #profit
            profit = (closing_price - opening_price) / opening_price * 100
            profits.append(profit)

            results.append({
                'Open Hour': hourly_df.iloc[i]['date'],
                'Open Price': opening_price,
                'Close Price': closing_price,
                'Profit': profit,
                'Stop Hit': stop_hit
            })

            in_trade = False

    results_df = pd.DataFrame(results)
    total_trades = len(results_df)
    successful_trades = len(results_df[results_df['Profit'] > 0])
    total_profit = sum(profits)
    profit_per_year = total_profit / (hourly_df['date'].dt.year.nunique())
    profit_per_trade = np.mean(profits) if profits else 0

    summary_df = pd.DataFrame({
        'Instrument': [name],
        'Number of trades': [total_trades],
        'Number of successful trades': [successful_trades],
        'Successful trades %': [100 * successful_trades / total_trades if total_trades > 0 else 0],
        'Profit %': [total_profit],
        'Profit % per year': [profit_per_year],
        'Profit % per trade': [profit_per_trade],

    })

    return summary_df

def strategy_evaluation_negative_cont(hourly_df, k=1.5, window=50, n=5, stop_loss=0.05, lookback=3, profit_goal=0.025, name='BTCUSD'):
    hourly_df = hourly_df.copy()
    hourly_df['hourly_returns'] = hourly_df['close'].pct_change()
    hourly_df.dropna(inplace=True)
    hourly_df['rolling_mean'] = hourly_df['hourly_returns'].rolling(window=window).mean() #rolling mean
    hourly_df['rolling_std'] = hourly_df['hourly_returns'].rolling(window=window).std() * np.sqrt(lookback) #depends on lookback

    #cumalitive returns
    hourly_df['cumulative_returns'] = hourly_df['hourly_returns'].rolling(window=lookback).sum()
    hourly_df['threshold'] = hourly_df['rolling_mean'] - k * hourly_df['rolling_std'] #threshold
    
    results = []
    profits = []
    in_trade = False  # THis shows that we are in trade or not so not overlap trade

    for i in range(len(hourly_df) - n):
        if in_trade:
            continue

        if hourly_df.iloc[i]['cumulative_returns'] < hourly_df.iloc[i]['threshold']:  
            opening_price = hourly_df.iloc[i]['close']
            closing_price = opening_price
            stop_hit = False

            # Enter trade
            in_trade = True

            for j in range(1, n + 1):
                current_price = hourly_df.iloc[i + j]['close']
                if (current_price - opening_price) / opening_price <= -stop_loss:  #end trade for stoploss
                    closing_price = current_price
                    stop_hit = True
                    break
                if (current_price - opening_price) / opening_price >= profit_goal:  #Profit goal condition
                    closing_price = current_price
                    break
                closing_price = current_price #else end trade after n
            
            #profit
            profit = (closing_price - opening_price) / opening_price * 100
            profits.append(profit)

            results.append({
                'Open Hour': hourly_df.iloc[i]['date'],
                'Open Price': opening_price,
                'Close Price': closing_price,
                'Profit': profit,
                'Stop Hit': stop_hit
            })

            in_trade = False

    results_df = pd.DataFrame(results)
    total_trades = len(results_df)
    successful_trades = len(results_df[results_df['Profit'] > 0])
    total_profit = sum(profits)
    profit_per_year = total_profit / (hourly_df['date'].dt.year.nunique())
    profit_per_trade = np.mean(profits) if profits else 0

    summary_df = pd.DataFrame({
        'Instrument': [name],
        'Number of trades': [total_trades],
        'Number of successful trades': [successful_trades],
        'Successful trades %': [100 * successful_trades / total_trades if total_trades > 0 else 0],
        'Profit %': [total_profit],
        'Profit % per year': [profit_per_year],
        'Profit % per trade': [profit_per_trade],

    })

    return summary_df

def strategy_evaluation_negative_mom(hourly_df, k=1.5, window=50, n=5, stop_loss=0.05, lookback=3, profit_goal=0.025, name='BTCUSD'):
    hourly_df = hourly_df.copy()
    hourly_df['hourly_returns'] = hourly_df['close'].pct_change()
    hourly_df.dropna(inplace=True)
    hourly_df['rolling_mean'] = hourly_df['hourly_returns'].rolling(window=window).mean() #rolling mean
    hourly_df['rolling_std'] = hourly_df['hourly_returns'].rolling(window=window).std() * np.sqrt(lookback) #depends on lookback

    #cumalitive returns
    hourly_df['cumulative_returns'] = hourly_df['hourly_returns'].rolling(window=lookback).sum()
    hourly_df['threshold'] = hourly_df['rolling_mean'] - k * hourly_df['rolling_std'] #threshold
    
    results = []
    profits = []
    in_trade = False  # THis shows that we are in trade or not so not overlap trade

    for i in range(len(hourly_df) - n):
        if in_trade:
            continue

        if hourly_df.iloc[i]['cumulative_returns'] < hourly_df.iloc[i]['threshold']:  
            opening_price = hourly_df.iloc[i]['close']
            closing_price = opening_price
            stop_hit = False

            # Enter trade
            in_trade = True

            for j in range(1, n + 1):
                current_price = hourly_df.iloc[i + j]['close']
                if (closing_price - current_price) / closing_price >= stop_loss:  #end trade for stoploss
                    closing_price = current_price
                    stop_hit = True
                    break
                if (opening_price - current_price) / opening_price >= profit_goal:  #Profit goal condition
                    closing_price = current_price
                    break
                closing_price = current_price #else end trade after n
            
            #profit
            profit = (opening_price - closing_price) / opening_price * 100
            profits.append(profit)

            results.append({
                'Open Hour': hourly_df.iloc[i]['date'],
                'Open Price': opening_price,
                'Close Price': closing_price,
                'Profit': profit,
                'Stop Hit': stop_hit
            })

            in_trade = False

    results_df = pd.DataFrame(results)
    total_trades = len(results_df)
    successful_trades = len(results_df[results_df['Profit'] > 0])
    total_profit = sum(profits)
    profit_per_year = total_profit / (hourly_df['date'].dt.year.nunique())
    profit_per_trade = np.mean(profits) if profits else 0

    summary_df = pd.DataFrame({
        'Instrument': [name],
        'Number of trades': [total_trades],
        'Number of successful trades': [successful_trades],
        'Successful trades %': [100 * successful_trades / total_trades if total_trades > 0 else 0],
        'Profit %': [total_profit],
        'Profit % per year': [profit_per_year],
        'Profit % per trade': [profit_per_trade],

    })

    return summary_df

```
